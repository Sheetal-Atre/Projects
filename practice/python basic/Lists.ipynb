```python

def find_kth_smallest_no(input_list, k):
    if(len(input_list)==0):
        return None
    else:
        #loop over till you find the k-1th smallest elemt 
        for i in range(0,k-1): 
            #find the min elmt in the list
            elmt = min(input_list)
            
            #list comprehension to remove all occurrences of the min elmt
            input_list = [item for item in input_list if item != elmt]   
            
        #return the last min no remaining in the list, which is kth smallest no
        if len(input_list) > 0:
            elmt = min(input_list)
        else:
            elmt = 'not present'
    return elmt

print(find_kth_smallest_no([2, -3, 0, 7, 21],2))#output = 0
print(find_kth_smallest_no([2, -3, 0, 7, 21],3))#output = 2
print(find_kth_smallest_no([-3, -3, 0, 7, 21],3))#output = 7
print(find_kth_smallest_no([6, 6, 6, 6, 6],2))#output = not present
```

Lego Stack
Description

You are given a row of Lego Blocks consisting of n blocks. All the blocks given have a square base whose side length is known. You need to stack the blocks over each other and create a vertical tower. Block-1 can go over Block-2 only if sideLength(Block-2)=>sideLength(Block-1).

From the row of Lego blocks, you can only pick up either the leftmost or rightmost block.

Print "Possible" if it is possible to stack all n cubes this way or else print "Impossible".


Input Format:

The input will contain a list of n integers representing the side length of each block's base in the row starting from the leftmost.


Sample Input:

[5 ,4, 2, 1, 4 ,5]

Sample Output:

Possible


```python
import ast,sys
#input_str = sys.stdin.read()
#sides = ast.literal_eval(input_str)#list of side lengths
sides=[5 ,4, 2, 1]

while len(sides) > 0:
    block2 = sides.pop()#right pop
    block1 = sides.pop(0)#left pop
    if block2 < block1:
        break
        
if len(sides) ==0:
    print("Possible")
else:
    print("Impossible")
   
```

```python

def pascal_triangle(lines):
    #2 d matrix of lines of pascal triangle
    triangle_matrix=[]*lines
    
    #append the first line, which is assumed. we process from 2nd line onwards.
    triangle_matrix.append([1])
    
    #init original list
    l=[]
    
    #process line 2 onwards
    for i in range(1,lines):
        
        #new list
        newl=[]
        
        #append 1 at start
        newl.append(1)

        #add all the elmt pairs in the prev line, and append each sum 
        #one by one to the new line list
        elmts_prev_list=len(l)
        for i in range(0,elmts_prev_list-1):
            s = l[i]+l[i+1]
            newl.append(s)

        #append 1 at end
        newl.append(1)

        #add the line to the matrix
        triangle_matrix.append(newl)
        
        #deep copy - set the prev line to the newly created line
        #now we can process next line
        l = list.copy(newl)
        
    return triangle_matrix

def print_triangle(matrix):
    for i in matrix:
        print(i)
        
def print_triangle_at(matrix, line):
    print(matrix[line-1])
    

lines_of_pascal_triangle=6
triangle_matrix = pascal_triangle(lines_of_pascal_triangle)
print_triangle(triangle_matrix)
print_triangle_at(triangle_matrix, lines_of_pascal_triangle)
```
